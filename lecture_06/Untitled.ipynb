```python
class Car:
    
    def __init__(self):
        self.__doors = 2
        self.wheels = 4
        
    def running(self):
        print("Running on wheels", self.__doors)
```

```python
class WonderCar(Car):
    
    def __init__(self):
        super().__init__()
        self.wheels = 2
        
    def running(self):
        print("Running on wheels", self.__doors)
```

```python
car = Car()
```

```python
car.wheels
```

```python
wonder = WonderCar()
```

```python
wonder.running()
```

```python
class Earth:
    
    __instance = None
    
    def __init__(self):
        self.people = []
    
    @classmethod
    def getInstance(cls):
        if cls.__instance == None:
            cls.__instance = Earth()
        
        return cls.__instance
```
