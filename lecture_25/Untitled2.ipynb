```python
X = np.linspace(-10, 10)

plt.figure()
plt.subplot(2, 2, 1)
plt.plot(X, np.sin(X))

plt.subplot(2, 2, 3)
plt.plot(X, np.cos(X))

plt.subplot(1, 2, 2)
plt.plot(X, np.tan(X))
plt.show()
```
